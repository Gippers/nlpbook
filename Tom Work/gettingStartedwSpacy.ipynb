```python
#Import spacy and download language model
import spacy
nlp = spacy.load("en_core_web_sm")
#This is the english language statistical model
```

```python
#tokenisation
sentence = nlp.tokenizer("We live in Paris.")

#length of sentence
print("The nmber of tokens: ", len(sentence))

#print individual works (i.e., tokens)
print("The tokens: ")
for words in sentence:
    print(words)
    
```

```python
#import libaries to Load in publically available jeopardy data
import pandas as pd
import os
cwd = os.getcwd()
```

```python
#Import Jeopardy Questions
data = pd.read_csv('C:/Users/thoma/data/jeopardy_questions/jeopardy_questions.csv')
data = pd.DataFrame(data=data)

#lowecase, strip whistespace ad view column names
data.columns = map(lambda x: x.lower().strip(), data.columns)
'''
    map(funciton, iterable) - takes a function and applies it to each iterable
    lambda arguments: exressions - small anonymous function. Takes multiple arguments but only one expression
    pandas.apply - apply a function along an axis of a dataframe. 
'''

#Reduce size of data
data = data[0:1000]

#tokenize Jeopardy Questions
data["question_tokens"] = data["question"].apply(lambda x: nlp(x))
print(data.head())
```

```python
#lets makes sure that this worked correctly.
#View first question
example_question = data.question[0]
example_question_tokens = data.question_tokens[0]
print("The first questions is:")
print(example_question)

#print individual tokens of the first questions
print("The tokens from the first question are:")
for tokens in example_question_tokens:
    print(tokens)
```

```python
'''
    Part-of speech tagging
'''
#print part-of-speech tags for tokens in the first question
print("Here are the Part-of-speech tags for eac token in the first questions:")
for token in example_question_tokens:
    print(token.text,token.pos_, spacy.explain(token.pos_))
```

```python
'''
    Dependency Parsing - Is the process of finding relationships among the tokens
    Essential step in NLP. The graph in particular shows how certain tokens and be grouped together. 
    Also demonstrates the importance of certain tokens. 
'''
#Print dependency parsing tags for tokens in the first question
for token in example_question_tokens:
    print(token.text, token.dep_, spacy.explain(token.dep_))
    
#visualise the dependency parse
from spacy import displacy

displacy.render(example_question_tokens, style='dep', jupyter=True, options={'distance':120})
```
